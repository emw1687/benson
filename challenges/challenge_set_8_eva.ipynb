### Challenge 1

For the house representatives data set, calculate the accuracy, precision, recall and f1 scores of each classifier you built (on the test set).

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X = votes
y = X.pop(16)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=4444)
```

```python
#load models from Challenge Set 7
logreg = open_pickle('data/logreg.pkl')
knn = open_pickle('data/knn.pkl')
nb = open_pickle('data/nb.pkl')
svm = open_pickle('data/svm.pkl')
dt = open_pickle('data/decisiontree.pkl')
rf = open_pickle('data/randomforest.pkl')
```

```python
classifiers = {'logreg': logreg,
               'knn': knn,
               'nb': nb,
               'svm': svm,
               'decisiontree': dt,
               'randomforest': rf}
```

```python
for name, classifier in classifiers.items():
    print name
    y_pred = classifier.predict(X_test)
    print 'accuracy:', metrics.accuracy_score(y_test, y_pred)
    precision, recall, fscore, support = metrics.precision_recall_fscore_support(y_test, y_pred, pos_label='democrat', average='binary')
    print 'precision:', precision
    print 'recall:', recall
    print 'fscore', fscore
    print '--------------------'
```

### Challenge 2

For each, draw the ROC curve and calculate the AUC.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
def plot_roc(modelname, title, pos_label):
    y_score = modelname.predict_proba(X_test)[:,0]
    fpr, tpr, thresholds = roc_curve(y_test, y_score, pos_label=pos_label)
    AUC = auc(fpr, tpr)
    fig_title = title + ', AUC:' + str(AUC)
    fig, ax = plt.subplots()
    fig = sns.regplot(fpr, tpr, fit_reg=False)
    fig.set(xlabel='FPR',ylabel='TPR')
    fig.axes.set_title(fig_title);
```

```python
for title, classifier in classifiers.items():
    plot_roc(classifier, title, pos_label='democrat')
```
